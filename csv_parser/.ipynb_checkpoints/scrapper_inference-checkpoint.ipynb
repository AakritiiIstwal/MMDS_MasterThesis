```python
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
scrapper_data = pl.read_csv('scrapper_dataset.csv')
print(scrapper_data)
```

```python
from ast import literal_eval

# Function to safely evaluate a string as a list
def safe_literal_eval(s):
    try:
        # Use literal_eval to safely evaluate the string to a list
        return literal_eval(s)
    except ValueError:
        # Return an empty list or some default value if the string is not a valid list
        return s
```

```python
#Plot Graphs
def plot_event_counts(event_counts):
    plt.figure(figsize=(4, 4))
    
    # Create a barplot and use a color palette that is visually appealing
    barplot = sns.barplot(
        data=event_counts, 
        x="events_label_list", 
        y="count", 
        palette="viridis"
    )
    
    # Rotate the x-labels for better readability
    plt.xticks(rotation=45, ha='right')
    
    # Optionally, you can tighten the layout to fit everything nicely
    plt.tight_layout()
    
    # Set the title and labels with a larger font for readability
    plt.title('Frequency of Event Labels', fontsize=18)
    plt.xlabel('Event Labels', fontsize=14)
    plt.ylabel('Count', fontsize=14)
    
    # Show the plot
    plt.show()
```

```python
for name, data in scrapper_data.group_by(['acoustic_scene_label'], maintain_order=True):
    print("Group name is : ", name[0])
    # print(data)
    # print(data.select(['acoustic_scene_label','events_label_list']))
    
    # Apply the conversion function to the events_label_list column
    corrected_event_col = data.select(pl.col("events_label_list").apply(safe_literal_eval).alias("events_label_list"))
    corrected_event_col = corrected_event_col.select(pl.col("events_label_list").list.explode())
    event_counts = (corrected_event_col
                    .groupby("events_label_list")
                    .agg(pl.count().alias('count'))
                    .sort(by="count", descending=True))    
    print(event_counts)
    plot_event_counts(event_counts)
    
```
