```python
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tut18_data = pl.read_csv('TUT18_train_panns_eventsInference.csv')
```

```python
tut18_data
```

```python
from ast import literal_eval

def safe_literal_eval(s):
    try:
        return literal_eval(s)
    except ValueError:
        return s
```

```python
#Plot Graphs
def plot_event_counts(event_counts, scene_label):

    plt.figure(figsize=(10, 10))
    bars = plt.barh(range(len(event_counts['events_label_list'])), event_counts['count'], color='skyblue')
    plt.yticks(range(len(event_counts['events_label_list'])), event_counts['events_label_list'], fontsize=8)
    plt.xlabel('Number of Occurrences')
    plt.title(f'Frequency of Events at {scene_label}')
    
    for bar in bars:
        plt.text(
            bar.get_width(),  # get the horizontal position of the text
            bar.get_y() + bar.get_height() / 2,  # get the vertical position of the text
            f'{bar.get_width():.0f}',  # round the number to 0 decimal places as text
            va='center',  # vertical alignment is center
            ha='left'  # horizontal alignment is to the left (so the text is after the bar)
        )
    
    plt.tight_layout()  # adjust subplot params so that the subplot(s) fits into the figure area
    plt.show()
```

```python
for name, data in tut18_data.group_by(['labels'], maintain_order=True):
    
    corrected_event_col = data.select(pl.col('panns_inference_events_list').apply(safe_literal_eval).alias("events_label_list"))
    corrected_event_col = corrected_event_col.select(pl.col("events_label_list").list.explode())
    event_counts = (corrected_event_col
                    .groupby("events_label_list")
                    .agg(pl.count().alias('count'))
                    .sort(by="count", descending=True))
    
    plot_event_counts(event_counts, name[0])
```
